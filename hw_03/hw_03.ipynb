```python
import pandas as pd
```

```python
train = pd.read_csv('03_santander_train.csv',  error_bad_lines=False,index_col='ID')
```

```python
train.head()
```

```python
train.info(verbose='true',null_counts='true')
```

```python
train = train[~train['TARGET'].isnull()]  #удалили пропуски(строки где TARGET пропущен)
train.shape
```

```python
test = pd.read_csv('03_santander_test.csv', index_col='ID')
```

```python
test.head()
```

```python
test.info(verbose='true',null_counts='true')
```

```python
y = list(train['TARGET'])
```

```python
train = train.drop('TARGET', axis=1)
```

```python
train.info(verbose=True, memory_usage=True, null_counts=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
```

```python
auc_lst_dtc=[]
for i in range(1,11):
    dtc = DecisionTreeClassifier(max_depth=i)
    dtc.fit(X_train, y_train)
    auc_lst_dtc.append(roc_auc_score(y_test, dtc.predict_proba(X_test)[:,1]))
    
auc_lst_dtc  
```

```python
df = pd.Series(auc_lst_dtc, [x for x in range(1,11)])
df
```

```python
from matplotlib import pyplot as plt
_, ax = plt.subplots(figsize=(20,10))
plt.ylabel('roc_auc_score')
plt.xlabel('max_depth')
plt.title('roc_auc_score_dtc of max_depth')
df.plot(kind='bar', ax=ax)  
plt.show()
```

```python
auc_lst_LR_L1 = []
for i in [0.1,1.0,10.0,100.0]:
    lr = LogisticRegression(C=i,penalty='l1')
    lr.fit(X_train, y_train)
    auc_lst_LR_L1.append(roc_auc_score(y_test,lr.predict_proba(X_test)[:,1]))
    
auc_lst_LR_L1 

 
```

```python
auc_lst_LR_L2 = []
for i in [0.1,1.0,10.0,100.0]:
    lr = LogisticRegression(C=i,penalty='l2')
    lr.fit(X_train, y_train)
    auc_lst_LR_L2.append(roc_auc_score(y_test,lr.predict_proba(X_test)[:,1]))
    
auc_lst_LR_L2 
```

```python
df = pd.DataFrame(index = [0.1,1.0,10.0,100.0], data = {'L1':auc_lst_LR_L1,'L2':auc_lst_LR_L2})
df
```

```python
from matplotlib import pyplot as plt
_, ax = plt.subplots(figsize=(20,10))
plt.ylabel('roc_auc_score')
plt.xlabel('C')
plt.title('roc_auc_score_lr of C')
df.plot(kind='bar', ax=ax)  
plt.show()
```

```python
auc_lst_knn=[]
for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    auc_lst_knn.append(roc_auc_score(y_test, knn.predict_proba(X_test)[:,1]))
auc_lst_knn 
```

```python
df = pd.Series(auc_lst_knn, [x for x in range(1,11)])
df
```

```python
from matplotlib import pyplot as plt
_, ax = plt.subplots(figsize=(20,10))
plt.ylabel('roc_auc_score')
plt.xlabel('n_neighbors')
plt.title('roc_auc_score_knn of n_neighbors')
df.plot(kind='bar', ax=ax)  
plt.show()
```
